## Q7. Different Job types and their pay distribution

```python
salary_with_pos = pd.DataFrame(IT_salary_data.groupby('Position ').median()['Yearly brutto salary (without bonus and stocks) in EUR'])
```

```python
top_salary_person = salary_with_pos.sort_values('Yearly brutto salary (without bonus and stocks) in EUR',ascending=False).astype(np.int64).reset_index().head(10)
top_salary_person.head()
```

```python
top_salary = top_salary_person.rename(columns = {'Position ':"Position","Yearly brutto salary (without bonus and stocks) in EUR":"Basic_salary"})
```

```python

fig = plt.figure(figsize =(20, 10))
 
# Horizontal Bar Plot
plt.bar(top_salary.Position, top_salary.Basic_salary, color="black",  edgecolor='red',linewidth = 4)
 
# Show Plot
# Show Plot

plt.grid(True)
plt.show()
```

## Q8. Distribution of salary vs programming language




```python
domain_wise_salary = IT_salary_data.groupby('Your main technology / programming language').mean()['Yearly brutto salary (without bonus and stocks) in EUR']
```

```python
domain_wise_salary_df = pd.DataFrame(domain_wise_salary).reset_index()
domain_wise_salary_df = domain_wise_salary_df.rename(columns = {'Your main technology / programming language':"Programming_lang","Yearly brutto salary (without bonus and stocks) in EUR":"basic_salary"})
```

```python
top_05_domain_wise_salary = domain_wise_salary_df.sort_values("basic_salary").head(5)
```

```python
fig = plt.figure(figsize =(25, 10))
 
# Horizontal Bar Plot
ax = plt.bar(top_05_domain_wise_salary.Programming_lang, top_05_domain_wise_salary.basic_salary,color="black",  edgecolor='red',linewidth = 4)
 
# Show Plot
plt.grid(True)
plt.xlabel('Programing language')
plt.ylabel('Salary in Euro')
plt.title('Programing language vs Salary ')
plt.xticks(rotation=20)
plt.show()
```
