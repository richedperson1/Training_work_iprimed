```python
import math
import scipy.stats as st
mu = 0
var = 100

std_var = math.sqrt(var)

x1 = np.linspace(mu-4*std_var,mu+4*std_var,100)
x2 = np.linspace(mu-4*5,mu+4*5,100)
x3 = np.linspace(mu-4*1,mu+4*1,100)
```

```python
plt.plot(x1,st.norm.pdf(x1,0,std_var))
plt.plot(x2,st.norm.pdf(x2,0,5))
plt.plot(x3,st.norm.pdf(x3,0,1))
```
